```python
import zipfile
import os

# Path to your zip file
zip_file_path = '/content/drive/MyDrive/w266/266 Final Project/Music4all Dataset/music4all.zip'
# Directory where you want to extract the files
extract_dir = '/content/drive/MyDrive/w266/266 Final Project/Music4all Dataset/unzip'
zip_password = 'ifjlryvm2019'

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    for file_info in zip_ref.infolist():
        if file_info.filename.startswith('music4all/lyrics/'):
            zip_ref.extract(file_info, extract_dir, pwd=bytes(zip_password, 'utf-8'))
```
